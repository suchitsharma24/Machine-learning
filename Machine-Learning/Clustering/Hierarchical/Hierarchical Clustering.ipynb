```python
#Import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Importing the dataset
dataset = pd.read_csv('data/Mall_Customers.csv')
```

```python
#fetch the columns (area of interest)
X = dataset.iloc[:,[3,4]].values
```

```python
#Dendrogram method to find optimal number of clusters for the dataset
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Dendrogram method')
plt.xlabel('Number of Clusters')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
#Apply Agglomerative Clustering for optimal number of cluster
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
y_hc = hc.fit_predict(X)
```

```python
#Visualizing the clusters
plt.scatter(X[y_hc==0,0], X[y_hc==0,1], s=100, color='red', label='Careful')
plt.scatter(X[y_hc==1,0], X[y_hc==1,1], s=100, color='green', label='Standard')
plt.scatter(X[y_hc==2,0], X[y_hc==2,1], s=100, color='blue', label='Target')
plt.scatter(X[y_hc==3,0], X[y_hc==3,1], s=100, color='cyan', label='Careless')
plt.scatter(X[y_hc==4,0], X[y_hc==4,1], s=100, color='magenta', label='Sensible')
plt.title('Client\'s cluster')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
